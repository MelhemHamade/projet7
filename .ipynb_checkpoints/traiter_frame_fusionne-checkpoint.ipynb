```python
import pandas as pd

def presence_and_variability(data):
    # Initialiser un DataFrame pour stocker les résultats
    results = pd.DataFrame(index=data.columns)
    
    # Calculer le pourcentage de présence des données pour chaque colonne
    results['Presence Rate (%)'] = (1 - data.isnull().mean()) * 100
    
    # Initialiser des colonnes pour la variabilité
    results['Standard Deviation'] = pd.NA
    results['Coefficient of Variation'] = pd.NA
    results['Most Common Category Frequency'] = pd.NA
    results['Unique Categories'] = pd.NA
    
    # Calculer les mesures de variabilité pour les variables quantitatives et catégorielles
    for col in data.columns:
        if pd.api.types.is_numeric_dtype(data[col]):
            # Pour les variables quantitatives
            results.at[col, 'Standard Deviation'] = data[col].std()
            mean = data[col].mean()
            results.at[col, 'Coefficient of Variation'] = data[col].std() / mean if mean != 0 else pd.NA
        else:
            # Pour les variables catégorielles
            most_common_freq = data[col].value_counts(normalize=True).max() * 100  # fréquence de la catégorie la plus commune
            results.at[col, 'Most Common Category Frequency'] = most_common_freq
            results.at[col, 'Unique Categories'] = data[col].nunique()
    
    # Trier les résultats par taux de présence décroissant
    results_sorted = results.sort_values(by='Presence Rate (%)', ascending=False)
    
    return results_sorted.round(2)

# Appliquer la fonction au DataFrame (remplacer 'your_dataframe' par votre DataFrame)
# results = presence_and_variability(your_dataframe)
# print(results)


results = presence_and_variability(train)
results
```

```python
def finite_unique_values_variables(df):

    # Créer un dictionnaire pour stocker le nombre de valeurs uniques pour chaque variable
    unique_counts = {col: df[col].nunique() for col in df.columns}
    
    # Filtrer pour obtenir les variables avec un nombre fini de valeurs uniques
    # Vous pouvez définir un seuil spécifique si nécessaire, par exemple, moins de 20 valeurs uniques
    finite_vars = {col: count for col, count in unique_counts.items() if count < 20}  # Exemple de seuil: 20
    
    # Afficher la liste des variables à nombre fini de valeurs uniques
    print(finite_vars)

    return finite_vars
```

```python
def remove_high_corr_vars(train, high_corr_file):
    high_corr_pairs = pd.read_csv(high_corr_file, header=None, index_col=[0, 1])
    high_corr_pairs.index.names = ['Var1', 'Var2']

    for var1, var2 in high_corr_pairs.index:
        # Vérifier si les deux variables sont toujours dans le DataFrame
        if var1 in train.columns and var2 in train.columns:
            # Compter les occurrences non nulles pour chaque variable
            count_var1 = (train[var1] != 0).sum()
            count_var2 = (train[var2] != 0).sum()

            # Supprimer la variable la moins présente
            if count_var1 < count_var2:
                train.drop(var1, axis=1, inplace=True)
            else:
                train.drop(var2, axis=1, inplace=True)

    return train

# Appliquer la fonction pour supprimer les variables de 'train'
train_modified = remove_high_corr_vars(train, "high_correlation_pairs.csv")
```
