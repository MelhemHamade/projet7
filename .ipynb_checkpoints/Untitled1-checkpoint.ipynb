```python
import pandas as pd

# Charger les donnÃ©es
train = pd.read_csv('application_train.csv')
test = pd.read_csv('application_test.csv')
bureau = pd.read_csv('bureau.csv')

# Combiner les identifiants de train et test
all_ids = pd.concat([train['SK_ID_CURR'], test['SK_ID_CURR']]).unique()

# Filtrer bureau pour les entrÃ©es correspondant Ã  SK_ID_CURR dans train ou test
bureau_filtered = bureau[bureau['SK_ID_CURR'].isin(all_ids)]

# Aggregations
agg_bureau = bureau_filtered.groupby('SK_ID_CURR').agg({
    'DAYS_CREDIT': ['mean', 'min', 'max'],
    'CREDIT_DAY_OVERDUE': ['mean'],
    'DAYS_CREDIT_ENDDATE': ['min'],
    'DAYS_ENDDATE_FACT': ['mean', 'max'],
    'AMT_CREDIT_MAX_OVERDUE': ['mean'],
    'CNT_CREDIT_PROLONG': ['sum'],
    'AMT_CREDIT_SUM': ['sum'],
    'AMT_CREDIT_SUM_DEBT': ['sum'],
    'AMT_CREDIT_SUM_OVERDUE': ['mean'],
    'DAYS_CREDIT_UPDATE': ['mean'],
    'AMT_ANNUITY': ['mean']
})

# Renommer les colonnes
agg_bureau.columns = ['_'.join(col).strip() for col in agg_bureau.columns.values]

# Compter les diffÃ©rents statuts de crÃ©dit actif
credit_active_dummies = pd.get_dummies(bureau_filtered['CREDIT_ACTIVE'], prefix='ca')
credit_active_dummies = credit_active_dummies.mul(bureau_filtered['SK_ID_CURR'], axis=0)
credit_active_counts = credit_active_dummies.groupby('SK_ID_CURR').sum()

# Compter les types de crÃ©dit
credit_type_dummies = pd.get_dummies(bureau_filtered['CREDIT_TYPE'], prefix='ct')
credit_type_dummies = credit_type_dummies.mul(bureau_filtered['SK_ID_CURR'], axis=0)
credit_type_counts = credit_type_dummies.groupby('SK_ID_CURR').sum()

# Devise la plus frÃ©quente
most_frequent_currency = bureau_filtered.groupby('SK_ID_CURR')['CREDIT_CURRENCY'].agg(lambda x: x.value_counts().index[0])

# Fusionner toutes les informations
bureau_final = agg_bureau.join(credit_active_counts, on='SK_ID_CURR', how='left')
bureau_final = bureau_final.join(credit_type_counts, on='SK_ID_CURR', how='left')
bureau_final = bureau_final.join(most_frequent_currency, on='SK_ID_CURR', how='left')

# Fusionner avec train et test
train = train.join(bureau_final, on='SK_ID_CURR', how='left')
test = test.join(bureau_final, on='SK_ID_CURR', how='left')
```

```python
import pandas as pd

# Liste des DataFrames
datasets = [ train, bureau_balance, credit_card_balance, POS_CASH_balance,
            previous_application, sample_submission, HomeCredit_columns_description, installments_payments, bureau]

# Liste des noms des DataFrames
names = [ 'train', 'bureau_balance', 'credit_card_balance', 'POS_CASH_balance',
         'previous_application', 'sample_submission', 'HomeCredit_columns_description', 'installments_payments', 'bureau']

# CrÃ©er un dictionnaire pour stocker les noms de colonnes de chaque DataFrame
columns_dict = {}

# Remplir le dictionnaire avec les noms de colonnes de chaque DataFrame
for df, name in zip(datasets, names):
    columns_dict[name] = set(df.columns)

# Initialiser un dictionnaire pour stocker les datasets contenant chaque colonne commune
common_columns_datasets = {}

# Trouver les noms de colonnes communs Ã  au moins deux DataFrames et les datasets qui les contiennent
for name, columns in columns_dict.items():
    for column in columns:
        for other_name, other_columns in columns_dict.items():
            if column in other_columns:
                if column not in common_columns_datasets:
                    common_columns_datasets[column] = set()
                common_columns_datasets[column].update([name, other_name])

# Filtrer pour ne garder que les colonnes prÃ©sentes dans au moins deux datasets
common_columns_datasets = {col: datasets for col, datasets in common_columns_datasets.items() if len(datasets) > 1}

# Convertir le dictionnaire en DataFrame pour une meilleure visualisation
common_columns_df = pd.DataFrame(list(common_columns_datasets.items()), columns=['Common_Column', 'Datasets'])

# Afficher le DataFrame des noms de colonnes communs et des datasets qui les contiennent
common_columns_df
```

```python
def change_columns_names(data, columns, suffix):
    # Liste des noms de colonnes Ã  modifier
    
    
    # Ajouter le suffixe 'PREV' aux colonnes spÃ©cifiÃ©es dans previous_application
    for col in columns:
        if col in data.columns:
            data.rename(columns={col: f'{col}_'+suffix}, inplace=True)
    
    # Afficher les nouvelles colonnes pour vÃ©rification
    print(data.columns)
```

```python
import pandas as pd

def cat_variables_with_high_presence(data, threshold=90):
    """
    Retourne la liste des variables numÃ©riques dont le taux de prÃ©sence est supÃ©rieur au seuil donnÃ©.

    :param data: DataFrame contenant les donnÃ©es.
    :param threshold: Seuil de prÃ©sence en pourcentage.
    :return: Liste des noms des variables numÃ©riques avec un taux de prÃ©sence supÃ©rieur au seuil.
    """
    # Initialiser une liste pour stocker les noms des variables qui rÃ©pondent au critÃ¨re
    high_presence_vars = []

    # Calculer le taux de prÃ©sence pour chaque variable numÃ©rique
    for col in data.select_dtypes(exclude=[np.number]).columns:
        presence_rate = 100 * data[col].notnull().mean()
        if presence_rate >= threshold:
            high_presence_vars.append(col)

    return high_presence_vars
```

```python
import pandas as pd

def num_variables_with_high_presence(data, threshold=90):
    """
    Retourne la liste des variables numÃ©riques dont le taux de prÃ©sence est supÃ©rieur au seuil donnÃ©.

    :param data: DataFrame contenant les donnÃ©es.
    :param threshold: Seuil de prÃ©sence en pourcentage.
    :return: Liste des noms des variables numÃ©riques avec un taux de prÃ©sence supÃ©rieur au seuil.
    """
    # Initialiser une liste pour stocker les noms des variables qui rÃ©pondent au critÃ¨re
    high_presence_vars = []

    # Calculer le taux de prÃ©sence pour chaque variable numÃ©rique
    for col in data.select_dtypes(include=[np.number]).columns:
        presence_rate = 100 * data[col].notnull().mean()
        if presence_rate >= threshold:
            high_presence_vars.append(col)

    return high_presence_vars
```

```python
import pandas as pd
import numpy as np

def remove_outliers(data, variables_list, num_std=3):
    """
    Supprime les valeurs aberrantes des variables spÃ©cifiÃ©es dans un DataFrame.

    :param data: DataFrame contenant les donnÃ©es.
    :param variables_list: Liste des noms des variables Ã  nettoyer.
    :param num_std: Nombre d'Ã©carts-types utilisÃ©s pour dÃ©finir une valeur aberrante.
    :return: DataFrame avec les valeurs aberrantes supprimÃ©es.
    """
    for var in variables_list:
        if var in data.columns:
            # Calculer la moyenne et l'Ã©cart-type
            mean, std = data[var].mean(), data[var].std()
            
            # DÃ©finir les seuils pour les valeurs aberrantes
            lower_bound, upper_bound = mean - num_std * std, mean + num_std * std
            
            # Remplacer les valeurs aberrantes par NaN ou une autre valeur spÃ©cifique
            data[var] = np.where(
                (data[var] > lower_bound) & (data[var] < upper_bound),
                data[var],
                np.NaN  # ou une autre valeur de votre choix
            )
    return data

```

```python
# Exemple d'utilisation

threshold = 90  # DÃ©finir le seuil de prÃ©sence
variables_list = num_variables_with_high_presence(train, threshold)
cleaned_train = remove_outliers(train, variables_list,num_std=2)
cleaned_train
```

```python
import pandas as pd

def presence_rate(data):
    # Calculer le pourcentage de prÃ©sence des donnÃ©es pour chaque colonne
    presence_rate = (1 - data.isnull().mean()) * 100
    
    # Trier la sÃ©rie presence_rate par valeurs dÃ©croissantes
    sorted_presence_rate = presence_rate.sort_values(ascending=False)
    
    # Formater chaque pourcentage avec deux chiffres aprÃ¨s la virgule et ajouter le caractÃ¨re '%'
    formatted_presence_rate = sorted_presence_rate.apply(lambda x: f'{x:.2f} %')
    
    # Retourner le taux de prÃ©sence formatÃ© et triÃ©
    return formatted_presence_rate
```
